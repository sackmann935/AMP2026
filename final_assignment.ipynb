## Why Radar?

Radar sensors offer several advantages over LiDAR for autonomous driving:

| Property | LiDAR | Radar |
|----------|-------|-------|
| Weather robustness | Poor in rain/fog | Excellent |
| Velocity measurement | Indirect (multi-frame) | Direct (Doppler) |
| Cost | High | Low |
| Range | < 100m | 250m+ |
| Point density | High (~100k points) | Low (~1k points) |
| Angular resolution | High | Low |

### Radar-Specific Challenges

When working with radar data, be aware of these challenges:

1. **Sparsity**: Radar produces ~100x fewer points than LiDAR. Your network may need larger receptive fields or multi-frame accumulation.

2. **Noise**: Radar has more false detections (clutter) and position uncertainty. Consider filtering or learning to handle noise.

3. **Velocity ambiguity**: Doppler velocity is radial (toward/away from sensor), not the true object velocity vector.

4. **Height estimation**: Radar has poor elevation resolution. Objects may appear at incorrect heights.

5. **Multi-path reflections**: Radar signals can bounce off surfaces, creating ghost detections.

These challenges also present **opportunities for novel solutions** in your research paper!

### Radar Point Cloud Format

The View of Delft dataset provides radar point clouds with the following features:

| Channel | Description | Unit | Typical Range |
|---------|-------------|------|---------------|
| x | Position (forward) | meters | [0, 50+] |
| y | Position (lateral) | meters | [-25, 25] |
| z | Position (height) | meters | [-3, 2] |
| RCS | Radar Cross Section | dBsm | [-30, 30] |
| v_r | Radial (Doppler) Velocity | m/s | [-30, 30] |

**Key differences from LiDAR:**
- **5 channels** instead of 4 (x, y, z, RCS, velocity vs. x, y, z, intensity)
- **Much sparser**: ~500-2000 points vs. ~100,000 points for LiDAR
- **Direct velocity**: Doppler effect provides radial velocity measurement
- **RCS**: Related to object size and material reflectivity
- **Noisier**: More false detections and position uncertainty
